```python
#Варіант 33 (3)
import numpy as np

# 1. Створіть одновимірний масив (вектор) з парними числами від 2 до 20 та виведіть його значення.
vector = np.arange(2, 21, 2)
print("1. Вектор з парними числами від 2 до 20:")
print(vector)
```

```python
# 2. Створіть двовимірний масив (матрицю) розміром 3x3, заповніть його випадковими дійсними числами від 0 до 1 та виведіть його значення.
matrix = np.random.uniform(0, 1, size=(3, 3))
print("2. Матриця 3x3 з випадковими дійсними числами від 0 до 1:")
print(matrix)
```

```python
# 3. Створіть масив розміром 3x3, заповніть його випадковими натуральними числами від 1 до 20 та виведіть його значення.
matrix = np.random.randint(1, 21, size=(3, 3))
print("3. Матриця 3x3 з випадковими натуральними числами від 1 до 20:")
print(matrix)
```

```python
# 4. Створіть масив розміром 4x4, заповніть його випадковими дійсними числами в діапазоні від 0 до 1 та знайдіть мінімальний елемент.
matrix = np.random.uniform(0, 1, size=(4, 4))
print("4. Мінімальний елемент в матриці 4x4 з випадковими дійсними числами від 0 до 1:")
print(np.min(matrix))
```

```python
# 5. Створіть два одновимірних масиви розміром 7, заповніть їх випадковими цілими числами в діапазоні від 1 до 15 та виконайте на них поелементні операції додавання, віднімання та множення.
array1 = np.random.randint(1, 16, size=7)
array2 = np.random.randint(1, 16, size=7)
print("5. Поелементні операції на масивах:")
print("Додавання:", array1 + array2)
print("Віднімання:", array1 - array2)
print("Множення:", array1 * array2)
```

```python
# 6. Створіть дві матриці 3x3, заповніть їх довільними цілими числами від 3 до 30 та знайдіть їх добуток.
matrix1 = np.random.randint(3, 31, size=(3, 3))
matrix2 = np.random.randint(3, 31, size=(3, 3))
print("6. Добуток двох матриць 3x3 з довільними цілими числами від 3 до 30:")
print(np.dot(matrix1, matrix2))
```

```python
# 7. Створіть дві матриці розміром 4x2 та 2x5, заповніть їх випадковими цілими числами в діапазоні від 1 до 7 та перемножте їх між собою.
matrix1 = np.random.randint(1, 8, size=(4, 2))
matrix2 = np.random.randint(1, 8, size=(2, 5))
print("7. Добуток двох матриць розміром 4x2 та 2x5:")
print(np.dot(matrix1, matrix2))
```

```python
# 8. Створіть матрицю розміром 2x2, заповніть її випадковими цілими числами в діапазоні від 1 до 8 та знайдіть її обернену матрицю.
matrix = np.random.randint(1, 9, size=(2, 2))
print("8. Обернена матриця для 2x2 з випадковими цілими числами від 1 до 8:")
print(np.linalg.inv(matrix))
```

```python
# 9. Створіть матрицю розміром 2x3, заповніть її випадковими дійсними числами в діапазоні від 0 до 1 та транспонуйте її.
matrix = np.random.uniform(0, 1, size=(2, 3))
print("9. Транспонована матриця 2x3 з випадковими дійсними числами від 0 до 1:")
print(matrix.T)
```

```python
# 10. Створіть матрицю розміром 3x4 та вектор розміром 4x1, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть добуток матриці на вектор.
matrix = np.random.randint(1, 11, size=(3, 4))
vector = np.random.randint(1, 11, size=(4, 1))
print("10. Добуток матриці 3x4 на вектор 4x1:")
print(np.dot(matrix, vector))
```

```python
# 11. Створіть матрицю розміром 2x3 та вектор розміром 3x1, заповніть матрицю випадковими дійсними числами від 0 до 1, а вектор — випадковими цілими числами в діапазоні від 1 до 10, і знайдіть добуток матриці на вектор.
matrix = np.random.uniform(0, 1, size=(2, 3))
vector = np.random.randint(1, 11, size=(3, 1))
print("11. Добуток матриці 2x3 на вектор 3x1:")
print(np.dot(matrix, vector))
```

```python
# 12. Створіть дві матриці розміром 4x4, заповніть їх випадковими цілими числами в діапазоні від 1 до 10 та знайдіть їхню суму.
matrix1 = np.random.randint(1, 11, size=(4, 4))
matrix2 = np.random.randint(1, 11, size=(4, 4))
print("12. Сума двох матриць розміром 4x4 з випадковими цілими числами від 1 до 10:")
print(matrix1 + matrix2)
```

```python
# 13. Створіть дві матриці розміром 4x4, заповніть їх випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть їх добуток.
matrix1 = np.random.uniform(-1, 1, size=(4, 4))
matrix2 = np.random.uniform(-1, 1, size=(4, 4))
print("13. Добуток двох матриць 4x4 з випадковими дійсними числами в діапазоні від -1 до 1:")
print(np.dot(matrix1, matrix2))
```

```python
# 14. Створіть матрицю розміром 3x3, заповніть її випадковими дійсними числами в діапазоні від -1 до 1 та знайдіть суму елементів матриці.
matrix = np.random.uniform(-1, 1, size=(3, 3))
print("14. Сума елементів матриці розміром 3x3 з випадковими дійсними числами в діапазоні від -1 до 1:")
print(np.sum(matrix))
```

```python
# 15. Створіть матрицю розміром 5x5, заповніть її випадковими дійсними числами в діапазоні від -0.5 до 0.5 та знайдіть вектор-стовпчик, що містить суму елементів кожного рядка матриці.
matrix = np.random.uniform(-0.5, 0.5, size=(5, 5))
print("15. Вектор-стовпчик, що містить суму елементів кожного рядка матриці 5x5 з випадковими дійсними числами в діапазоні від -0.5 до 0.5:")
print(np.sum(matrix, axis=1).reshape(-1, 1))
```
