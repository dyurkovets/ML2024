```python
# Завантаження необхідних бібліотек
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
# 1. Завантаження вбудованого датасету
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = pd.Series(data.target)
```

```python
# 3. Розділення даних на навчальний і тестовий набори
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# 4. Побудова моделей
# Логістична регресія
lr = LogisticRegression(max_iter=10000)
# Дерево рішень
dt = DecisionTreeClassifier(random_state=42)
# Випадковий ліс
rf = RandomForestClassifier(random_state=42)
```

```python
# 5. Налаштування гіперпараметрів за допомогою GridSearchCV
# Параметри для пошуку
param_grid_lr = {'C': [0.01, 0.1, 1, 10], 'penalty': ['l2']}
param_grid_dt = {'max_depth': [3, 5, 10, None], 'min_samples_split': [2, 5, 10]}
param_grid_rf = {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20, 30]}

# GridSearch для кожної моделі
grid_lr = GridSearchCV(lr, param_grid_lr, cv=5)
grid_dt = GridSearchCV(dt, param_grid_dt, cv=5)
grid_rf = GridSearchCV(rf, param_grid_rf, cv=5)

# Навчання моделей
grid_lr.fit(X_train, y_train)
grid_dt.fit(X_train, y_train)
grid_rf.fit(X_train, y_train)
```

```python
# 6. Оцінка моделей
models = {'Logistic Regression': grid_lr, 'Decision Tree': grid_dt, 'Random Forest': grid_rf}

for name, model in models.items():
    print(f"Model: {name}")
    y_pred = model.predict(X_test)
    print(f"Accuracy: {accuracy_score(y_test, y_pred):.4f}")
    print(classification_report(y_test, y_pred))
    print(confusion_matrix(y_test, y_pred))
    print('-' * 50)
```

Random Forest і Logistic Regression показали найвищу точність (0.9708) на тестовій вибірці. Однак, Random Forest також продемонстрував трохи кращі результати в специфічності для класу 1, тому Random Forest є кращим вибором.Decision Tree має найнижчу точність.

```python
# Виведення найкращих параметрів для кожної моделі
print("\nНайкращі параметри для кожної моделі:")
best_params = {}
for name, model in models.items():
    best_params[name] = model.best_params_
    print(f"{name}: {model.best_params_}")
```
